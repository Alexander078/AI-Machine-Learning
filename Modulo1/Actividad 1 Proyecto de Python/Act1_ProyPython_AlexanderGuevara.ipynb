```python
# Realizando cambio de tipos
df_vehiculos['manufacturer_name'] = df_vehiculos['manufacturer_name'].astype('string')
df_vehiculos['model_name'] = df_vehiculos['model_name'].astype('string')
df_vehiculos['color'] = df_vehiculos['color'].astype('string')
df_vehiculos['engine_fuel'] = df_vehiculos['engine_fuel'].astype('string')
df_vehiculos['body_type'] = df_vehiculos['body_type'].astype('string')
df_vehiculos['location_region'] = df_vehiculos['location_region'].astype('string')
df_vehiculos['transmission'] = df_vehiculos['transmission'].astype('category')
df_vehiculos['engine_type'] = df_vehiculos['engine_type'].astype('category')
df_vehiculos['state'] = df_vehiculos['state'].astype('category')
df_vehiculos['drivetrain'] = df_vehiculos['drivetrain'].astype('category')
```

```python
#Verificando que se realizaron las conversiones, notar que la memoria usada se redujo a 4.4 MB en el dataframe
df_vehiculos.info()
```

```python
# Gestionar valores nulos o desconocidos
# verificando la cantidad de nulos por columnas
df_vehiculos.isnull().sum()
```

```python
# Mostrando las filas que tienen valores nulos, observamos que la columna engine_capacity tiene 10 valores nulos
df_vehiculos[df_vehiculos.isnull().any(axis=1)]
```

```python
#Visualizando los valores de la columna engine_capacity
df_vehiculos['engine_capacity'].head(5)
```

```python
#calculando el valor promedio de la columna engine_capacity
round(df_vehiculos['engine_capacity'].mean(),3)
```

```python
# obteniendo los indices de las filas que contienen nulos
values_index_null = df_vehiculos[df_vehiculos.isnull().any(axis=1)].index
```

```python
# sustituyendo los valores nulos de la columna engine_capacity por el promedio de los demas valores para no eliminar estas filas del dataset
df_vehiculos.loc[values_index_null, 'engine_capacity']= round(df_vehiculos['engine_capacity'].mean(),3)
```

```python
# Comprobando las filas que contienen nulos nuevamente, observamos que ya tenemos nulos en el dataframe
df_vehiculos[df_vehiculos.isnull().any(axis=1)]
```

```python
# Gestionar valores duplicados
# verificando la cantidad de duplicados del dataframe al parecer tenemos 40 duplicados en el fichero
df_vehiculos.duplicated().sum()
```

## 4. Visualización de resultados

```python
# Visualizamos la distribucion de los datos en histogramas de las distintas variables
df_vehiculos.hist(bins=10,figsize=(9,7),grid=False);
```

```python
# Visualizamos la transmision de los vehiculos y observamos que el dataset tiene mas vehiculos con transmision mecanica
sns.countplot(x='transmission', data=df_vehiculos,  palette='viridis')
plt.title("Transmision de vehiculos")
plt.show()
```

```python
# Visualizamos el tipo de motor de los vehiculos y observamos que el dataset tiene mas vehiculos con tipo de motor gasolina
sns.countplot(x='engine_type', data=df_vehiculos, palette='viridis')
plt.title("Tipo de motor")
plt.show()
```

```python
# obtenemos numero de colores
num_colors = len(df_vehiculos['color'].value_counts())

# Crear la lista explode con 0 para todas las porciones excepto la primera.
explode = [0.05] + [0] * (num_colors - 1)

# visualizar el gráfico de pastel de los colores de los vehiculos, observamos que el color black es en su mayoria con un 20% del total de los vehiculos
df_vehiculos['color'].value_counts().plot.pie(autopct='%1.2f%%', explode=explode)
```

```python
# graficando la correlación lineal entre las variables
# observamos que hay una alta correlacion entre el año de produccion del vehiculo (year_produced) y el precio (price_usd)
# tambien hay una correlacion negativa entre el valor de distancia recorrida (odometer_value) y el precio (price_usd)
numeric_df = df_vehiculos.select_dtypes(include=np.number)
corr=numeric_df.corr()

plt.figure(figsize=(10, 10))


sns.heatmap(corr, vmax=.8, linewidths=0.01,
            square=True,annot=True,cmap='YlGnBu',linecolor="white")
plt.title('Correlation between features');
```

```python
# visualizando boxplot de los precios
ax = sns.boxplot(x="price_usd", y="transmission",
                data=df_vehiculos)

plt.title("Precio por transmission",fontsize=12);
```

```python
# Visualizando top 10 de cantidad de vehiculos por fabricante en el dataset
df_vehiculos.manufacturer_name.value_counts().head(10).plot(kind="bar", title ='Vehiculos por fabricante')
```
